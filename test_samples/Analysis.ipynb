```python
import pandas as pd
import numpy as np
import re
from statistics import mean
```

```python
spain_df = pd.read_csv('tweets_argentina.csv', '\t')
```

```python
tweets = list(spain_df['Tweets'])
```

```python
tweets
```

```python
mean(list(map(len, (s.split() for s in tweets))))
```

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```

```python
clean_tweets = list()
for i in tweets:
    x = remove_emojis(i)
    clean_tweets.append(x)
    
```

```python
mean(list(map(len, (s.split() for s in clean_tweets))))
```
